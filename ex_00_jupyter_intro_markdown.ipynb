# Wyróżnienia tekstu  <a name="Emphasis"></a>

- Pochylenie tekstu (otoczenie tekstu znakiem `*`)  *asterisks* or _underscores_.

- Pochylenie i pogrubienie (otoczenie tekstu znakiem `**`) **asterisks** or __underscores__.

- Mieszane wyróżnienia (otoczenie tekstu znakiem `__`)  **asterisks and _underscores_**.

- Przekreślenia (otoczenie tekstu znakiem `~~`):  ~~ Scratch this ~~


*This text will be italic*
_This will also be italic_

**This text will be bold**
__This will also be bold__

_You **can** combine them_


# Tabele <a name="Tabele"></a>


**Bardzo pomocne narzędzie do tworzenia tabeli:** https://www.tablesgenerator.com/markdown_tables

Tabele nie są częścią podstawowej specyfikacji Markdown, ale są częścią GFM i Markdown Here je obsługuje. Są łatwym sposobem dodawania tabel do wiadomości e-mail - zadanie, które w innym przypadku wymagałoby kopiowania i wklejania z innej aplikacji.

Składnia tabeli GFM Markdown jest dość prosta. Nie pozwala na dzielenie wierszy lub komórek, a także umieszczanie tekstu wielowierszowego w komórce. Pierwszy wiersz to zawsze nagłówek, po którym następuje dodatkowa linia z myślnikami „-” i opcjonalnymi dwukropkami „:” w celu wymuszenia wyrównania kolumn.

Symbol `|` (Colons) może być uzyty do wyrównania kolumny

| Tables   |      Are      |  Cool |
|:--------:|:-------------:|  ----:|
| col 1 is |  left-aligned |  1600 |
| col 2 is |    centered   |    12 |
| col 3 is | right-aligned |    11 |

Każda komórka nagłówka musi zawierać co najmniej 3 myślniki, dwukropki otaczające myślniki symbolizują wyrównanie. Zewnętrzne (|) są opcjonalne i nie ma konieczności ustawiania surowej wersji tabeli Markdown.

Markdown | Less | Pretty
--- | :---: | ---:
*Still* | `renders` | **nicely**
1 | 2 | 3

# Kod HTML w lini <a name="InlineHTML"></a>
 
Możesz także użyć surowego HTML w edytorze Markdown i będzie to działać całkiem dobrze. Przykład:
```
<dl>
  <dt>Definicja</dt>
  <dd>Coś bardzo użytecznego</dd>

  <dt>HTML w Markdown w</dt>
  <dd>Does *not* work **very** well. Use HTML <em>tags</em>.</dd>
</dl>
```

# Wzory matematyczne <a name="math"></a>
Składnia podobna do składni LaTeX:

Wzór w lini otaczamy pojedyńczym znakiem `$` np.  `$x = \frac{x}{y}$` pokaże $x = \frac{x}{y}$

Wzrór wyśrodkowany otaczamy podwójnym znakiem `$$`, wygląda to tak:
$$x = \frac{x}{y} - a^2 + \sqrt{a -b}$$
